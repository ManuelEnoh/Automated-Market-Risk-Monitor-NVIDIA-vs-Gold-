```python
# 1. Import des librairies
import yfinance as yf
import pandas as pd

# 2. D√©finition des actifs
# ^GSPC = S&P 500 (Le March√©)
# NVDA = Nvidia (La Tech/Volatilit√©)
# GLD = Gold (La valeur refuge)
tickers = ['^GSPC', 'NVDA', 'GLD']

print("üöÄ T√©l√©chargement des donn√©es en cours...")

# 3. T√©l√©chargement (depuis 2020 pour avoir le Covid et l'inflation)
# On ne garde que la colonne 'Close' (Prix de fermeture)
data = yf.download(tickers, start="2020-01-01")['Close']

# 4. Nettoyage et Formatage pour SQL
# yfinance donne un tableau large (une colonne par action).
# On veut un tableau long (Date | Ticker | Prix) pour SQL.
data = data.reset_index() # La date devient une colonne
data_melted = data.melt(id_vars=['Date'], var_name='Ticker', value_name='Close_Price')

# 5. Aper√ßu et Sauvegarde
print("\n‚úÖ Aper√ßu des donn√©es :")
print(data_melted.head())

# Sauvegarde en CSV sans l'index (pour √©viter une colonne inutile)
data_melted.to_csv('market_data_raw.csv', index=False)
print("\nüíæ Fichier 'market_data_raw.csv' g√©n√©r√© avec succ√®s !")
```
